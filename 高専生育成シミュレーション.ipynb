# **高専生シミュレーター**

## **概要**

高専を舞台とした育成シミュレーションゲームです。ストレス値に注意しながら、3つのパラメータをあげて、試験合格と進級を目指します。

## **操作方法**

最初のセルを実行すれば開始します。まず、名前を入力して、あとは半角の数字で操作をしてください。誤入力してもエラーは出ないと確認しました。

## **制作者のコメント**

このゲームは、「高専生、頑張りすぎ」という思いから作られたものです。プログラムにはGithub Copilotが生成したものを参考にしたところがあり、それでclassについて知ることができました。イベント発生時の選択一覧の表示のところのプログラムや、試験開始の時期をそのまま埋め込めず、いつでも増やしたり、減らしたりできるところが工夫したポイントです。逆に、パラメータが上がるときの処理や、パラメータが上がったという報告のところは、よりきれいな書き方があると思いました。

## **作成時間**

作成時間：**約10時間**

```python
import random
import time

class Player:
  def __init__(self, name, ambition, experience, coexistence, stress, credit, score):
    self.name = name
    self.ambition = ambition
    self.experience = experience
    self.coexistence = coexistence
    self.stress = stress
    self.credit = credit
    self.score = score

class Event:
  def __init__(self, message, choices, end, amb, exp, coe, stress, month):
    self.message = message
    self.choices = choices
    self.end = end
    self.amb = amb
    self.exp = exp
    self.coe = coe
    self.stress = stress
    self.month = month

class Exam:
  def __init__(self, title, quota, ace, waves, difficulty, bonus, stress, month):
    self.title = title
    self.quota = quota
    self.ace = ace
    self.waves = waves
    self.difficulty = difficulty
    self.bonus = bonus
    self.stress = stress
    self.month = month

player = Player("高専太郎", 100, 100, 100, 0, 0, 0)

olympics = Event(
    "高専オリンピックが開催されます。",
    ["リレーに挑戦する", "モルックに挑戦する", "卓球に挑戦する", "参加しない"],
    ["リレーをやってみて良かった！", "モルックで頭を使った！", "卓球でいろんな人と試合ができた！", "人生はのんびりするもんや！"],
    [150, 0, 50, 0],
    [50, 150, 0, 0],
    [0, 50, 150, 0],
    [20, 20, 20, -50],
    5
)
assaignment = Event(
    "課題で独自のプロジェクトを提出するようになりました。",
    ["授業に出てこなかったものを取り入れる", "詳しい友達にアドバイスをもらう"],
    ["面白いものができて、褒められた！", "友達にいろいろ教わって、楽しくできた！"],
    [100, 50],
    [300, 300],
    [50, 100],
    [50, 50],
    7
)
summer_break = Event(
    "夏休みがやってきました。",
    ["専門分野の知識を広める", "ものすごいプロジェクトを作り始める", "習い事に参加する", "自分のペースで夏休みを過ごす"],
    ["自分の分野への理解を深めることができた！", "大きな課題に挑戦でき、面白かった！", "いろんな人と交流でき、楽しかった！", "のんびりしつつ、勉強もできた！"],
    [500, 300, 150, 250],
    [150, 500, 300, 250],
    [300, 150, 500, 250],
    [50, 50, 50, 0],
    8
)
festival = Event(
    "高専祭が開催されます。",
    ["自分のクラスの出し物に注目する", "他のクラスの出し物を見に行く"],
    ["仲間と協力できて、楽しかった！", "出し物が面白かった！"],
    [200, 400],
    [80, 80],
    [400, 200],
    [0, 0],
    11
)
spring_break = Event(
    "春休みがやってきました。学期が終わる前でも頑張りましょう。",
    ["専門分野の知識を広める", "ものすごいプロジェクトを作り始める", "習い事に参加する"],
    ["自分の分野への理解を深めることができた！", "大きな課題に挑戦でき、面白かった！", "いろんな人と交流でき、楽しかった！"],
    [1000, 0, 0],
    [0, 1000, 0],
    [0, 0, 1000],
    [0, 0, 0],
    3
)

event_list = [olympics, assaignment, summer_break, festival, spring_break]

first_semester_mid_exam = Exam("前期中間試験", 900, 1200, 3, 1, [50, 50, 50], -50, 6)
first_semester_final_exam = Exam("前期期末試験", 900, 1200, 3, 2, [45, 45, 45], -100, 8)
second_semester_mid_exam = Exam("後期中間試験", 900, 1200, 3, 3, [40, 40, 40], -50, 12)
second_semester_final_exam = Exam("後期期末試験", 900, 1200, 3, 4, [35, 35, 35], -100, 2)

exam_list = [first_semester_mid_exam, first_semester_final_exam, second_semester_mid_exam, second_semester_final_exam]

def event(event):
  status()
  print("イベントが発生！")
  print(event.message)
  time.sleep(0.5)
  print("どうしますか？")
  for i in range(len(event.choices)):
    print(f"{i + 1}:{event.choices[i]}")
  while True:
    while True:
      try:
        choice = int(input(f"半角数字を入力してください:"))
        break
      except ValueError:
        pass
    if choice > 0 and choice <= len(event.choices):
      ambev = event.amb[choice - 1] - \
          abs(event.amb[choice - 1] * player.stress / 100)
      expev = event.exp[choice - 1] - \
          abs(event.exp[choice - 1] * player.stress / 100)
      coeev = event.coe[choice - 1] - \
          abs(event.coe[choice - 1] * player.stress / 100)
      player.ambition += int(ambev)
      player.experience += int(expev)
      player.coexistence += int(coeev)
      player.stress += event.stress[choice - 1]
      print(event.end[choice - 1])
      time.sleep(0.5)
      if ambev > 0:
        print(f"「進取」が{int(ambev)}上がった！")
      elif ambev < 0:
        print(f"「進取」が{int(ambev)}下がった！")
      if expev > 0:
        print(f"「実践」が{int(expev)}上がった！")
      elif expev < 0:
        print(f"「実践」が{int(expev)}下がった！")
      if coeev > 0:
        print(f"「共生」が{int(coeev)}上がった！")
      elif coeev < 0:
        print(f"「共生」が{int(coeev)}下がった！")
      if event.stress[choice - 1] > 0:
        print(f"「ストレス」が{event.stress[choice - 1]}上がった！")
      elif event.stress[choice - 1] < 0:
        print(f"「ストレス」が{event.stress[choice - 1]}下がった！")
      if player.stress < 0:
        player.stress = 0
      if player.stress >= 100:
        player.stress = 100
        print("ストレスが最大です！パラメータがあがりにくくなります。少しのんびりしましょう。")
      player.score -= player.stress
      time.sleep(0.5)
      break
    else:
      pass

def exam(exam):
  print("--------------------")
  print(exam.title)
  input("Enterキーを押して始めましょう。")
  print(f"目標点数は{exam.quota}点、最大{exam.ace}点。")
  time.sleep(0.5)
  score = 0
  for i in range(exam.waves):
    if exam.waves > 1:
      print(f"{i+1}回目")
    score += (player.ambition * exam.bonus[0] / 100) / exam.difficulty
    score += (player.experience * exam.bonus[1] / 100) / exam.difficulty
    score += (player.coexistence * exam.bonus[2] / 100) / exam.difficulty
    time.sleep(0.5)
    if score >= exam.ace:
      print("満点！")
      time.sleep(0.5)
      break
    else:
      remaining_quota = max(0, exam.quota - score)
      print(f"点数は{int(score)}だと予想、目標点数まで{int(remaining_quota)}。")
      time.sleep(0.5)
  if score >= exam.quota:
    print("いい結果を出せた！")
    time.sleep(0.5)
  elif exam.ace / 2 < score < exam.quota:
    print("まあまあの結果だった。")
    player.credit += random.choices([1, 2], weights=[(player.ambition + player.experience + player.coexistence) / exam.difficulty, sum(exam.bonus)])[0]
    time.sleep(0.5)
  else:
    print("残念な結果だった。")
    player.credit += random.choices([3, 4], weights=[(player.ambition + player.experience + player.coexistence) / exam.difficulty, sum(exam.bonus)])[0]
    time.sleep(0.5)
  player.stress += exam.stress
  player.score += score - player.stress
  if exam.stress > 0:
    print(f"「ストレス」が{exam.stress}上がった！")
  elif exam.stress < 0:
    print(f"「ストレス」が{exam.stress}下がった！")
  if player.stress < 0:
    player.stress = 0
  if player.stress >= 100:
    player.stress = 100
    print("ストレスが最大です！パラメータがあがりにくくなります。少しのんびりしましょう。")
  time.sleep(0.5)

def status():
  print("--------------------")
  print(f"{player.name}")
  print(f"「進取」:{player.ambition}")
  print(f"「実践」:{player.experience}")
  print(f"「共生」:{player.coexistence}")
  print(f"ストレス:{player.stress}")
  print(f"落とした単位:{player.credit}")
  print("--------------------")

print("高専生シミュレーター")
player.name = input("名前を入力してください:")
month = 4
bonus = 1
for i in range(12):
  ambval = 50 + int(bonus**2) - abs(int((50 + bonus**2 / 10) * player.stress / 100))
  expval = 50 + int(bonus**2) - abs(int((50 + bonus**2 / 10) * player.stress / 100))
  coeval = 50 + int(bonus**2) - abs(int((50 + bonus**2 / 10) * player.stress / 100))
  print("--------------------")
  print(f"{month}月")
  status()
  print("今月は何をしますか？")
  print("1:勉強に励む")
  print("2:自分のプロジェクトに取り組む")
  print("3:校内の人間関係を築く")
  print("4:のんびりする")
  print("--------------------")
  choice = input("半角数字を入力してください:")
  while True:
    if choice == "1":
      player.ambition += ambval
      player.stress += 20
      if ambval > 0:
        print(f"「進取」が{ambval}上がった！")
      print("ストレスが20上がった！")
      break
    elif choice == "2":
      player.experience += expval
      player.stress += 20
      if expval > 0:
        print(f"「実践」が{expval}上がった！")
      print("ストレスが20上がった！")
      break
    elif choice == "3":
      player.coexistence += coeval
      player.stress += 20
      if coeval > 0:
        print(f"「共生」が{coeval}上がった！")
      print("ストレスが20上がった！")
      break
    elif choice == "4":
      player.stress -= 30
      print("ストレスが30下がった！")
      break
    else:
      choice = input("1から4の半角数字を入力してください:")
  if player.stress < 0:
    player.stress = 0
  if player.stress >= 100:
    player.stress = 100
    print("ストレスが最大です！パラメータがあがりにくくなります。少しのんびりしましょう。")
  time.sleep(0.5)
  for exams in exam_list:
    if month == exams.month:
      exam(exams)
  for events in event_list:
    if month == events.month:
      event(events)
  player.score += player.ambition + player.experience + player.coexistence - player.stress
  month += 1
  bonus += 1
  if month > 12:
    month = 1
print("--------------------")
print("1年が終わりました！")
print(f"落とした単位数:{player.credit}")
if player.credit >= 6:
  print("単位を落としすぎてしまい、あなたは留年してしまいました・・・")
else:
  print("あなたは進級できました！")
print(f"あなたのスコアは{int(player.score)}です！")
print("ゲーム終了です！お疲れ様でした！")
```
